```python
#What are the total sales for each category?
sales_by_category = supermarket_df.groupby('Category')['Sales'].sum()
print(sales_by_category)
```

```python
sales_by_category = supermarket_df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Which sub-category has the highest sales?
sales_by_subcategory = supermarket_df.groupby('Sub-Category')['Sales'].sum()
highest_sales_subcategory = sales_by_subcategory.idxmax()
print(highest_sales_subcategory)
```

```python
sales_by_subcategory = supermarket_df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 7))
sales_by_subcategory.plot(kind='bar')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
#What is the average sales amount per order?
average_sales_per_order = supermarket_df.groupby('Order ID')['Sales'].sum().mean()
average_sales_per_order
```

```python
#How do sales vary across different ship modes?
sales_by_ship_mode = supermarket_df.groupby('Ship Mode')['Sales'].sum()
print(sales_by_ship_mode)
```

```python
sales_by_ship_mode = supermarket_df.groupby('Ship Mode')['Sales'].sum()
plt.figure(figsize=(8, 5))
sales_by_ship_mode.plot(kind='bar')
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#What is the distribution of sales across different regions?
sales_by_region = supermarket_df.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
sales_by_region = supermarket_df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
sales_by_region.plot(kind='bar')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#What are the top 10 customers by total spending
top_10_customers = supermarket_df.groupby('Customer Name')['Sales'].sum().nlargest(10)
print(top_10_customers)
```

```python
# Group by Category and Region and sum Sales
sales_by_category_region = supermarket_df.groupby(['Category', 'Region'])['Sales'].sum().unstack()

# Plotting the stacked column chart
sales_by_category_region.plot(kind='bar', stacked=True, figsize=(12, 7))

plt.title('Total Sales by Category and Region')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

```python
#How many unique customers are there?
num_categories = supermarket_df['Category'].nunique()
print(num_categories)
```

```python
##What is the average number of orders per customer?
avg_no_of_orders_per_customer = supermarket_df.groupby('Customer Name')['Order ID'].nunique().mean()
avg_no_of_orders_per_customer
```

```python
orders_per_customer = supermarket_df.groupby('Customer Name')['Order ID'].nunique(plt.figure(figsize=(10, 6)))
sns.histplot(orders_per_customer, bins=30, kde=True)
plt.title('Distribution of Orders per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.show()
```

```python
orders_per_customer = supermarket_df.groupby('Customer Name')['Order ID'].nunique()
plt.figure(figsize=(10, 6))
sns.histplot(orders_per_customer, bins=30, kde=True)
plt.title('Distribution of Orders per Customer')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#Which customer segment has the highest sales?
highest_sales_customer = supermarket_df.groupby('Customer Name')['Sales'].sum().idxmax()
print(highest_sales_customer)
```

```python
if 'Segment' in supermarket_df.columns:
      sales_by_segment = supermarket_df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
      plt.figure(figsize=(8, 5))
      sales_by_segment.plot(kind='bar')
      plt.title('Total Sales by Customer Segment')
      plt.xlabel('Customer Segment')
      plt.ylabel('Total Sales')
      plt.xticks(rotation=0)
      plt.tight_layout()
      plt.show()
else:
      print("Column 'Segment' not found in the DataFrame.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(supermarket_df['Sales'], bins=50, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#Which are the most frequently ordered products?
most_frequent_products = supermarket_df['Sub-Category'].value_counts()
print(most_frequent_products)
```

```python
#Are there any correlations between different features (e.g., sales and discount)?
correlation_matrix = supermarket_df.corr(numeric_only = True)
print(correlation_matrix)
```

```python
correlation_matrix = supermarket_df.corr(numeric_only = True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
